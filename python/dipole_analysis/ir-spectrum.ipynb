```python
import numpy as np
import matplotlib.pyplot as plt
import ir
from scipy.signal import correlate
```

```python
def read_dipole_data(filename, numlines=-1):
    with open(filename, mode='r') as f:
        lines = f.readlines()
        dipoles = np.zeros((len(lines), 3), dtype=np.float64)
        for i, line in enumerate(lines):
            x, y, z = line.split()
            dipoles[i,:] = x, y, z
        
        return dipoles
```

```python
dipoles = read_dipole_data('/Users/Spencer/Documents/Research/Markland/dftbplus/dipoles.dat')
```

```python
autocorr = ir.autocorrelation(dipoles.T)
```

```python
plt.plot(autocorr)
```

```python
freq, spectrum = ir.ir_spectrum(autocorr, 2e-15)
```

```python
plt.plot(freq, spectrum)
```
